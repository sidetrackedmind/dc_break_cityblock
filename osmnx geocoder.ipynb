```python
import osmnx as ox
ox.config(log_console=True, use_cache=True)
```

```python
# network from address, including only nodes within 1km along the network from the address
G = ox.graph_from_address(address='2701 14th St NW, Washington, DC, USA', dist=1000,
                          dist_type='network', network_type='drive')

# you can project the network to UTM (zone calculated automatically)
G_projected = ox.project_graph(G)
```

```python
fig, ax = ox.plot_graph(G_projected, node_size=0, edge_color='w', edge_linewidth=0.2)
```

```python
G2 = ox.consolidate_intersections(G_projected, rebuild_graph=True, tolerance=15, dead_ends=False)
```

```python
fig, ax = ox.plot_graph(G2, node_size=0, edge_color='w', edge_linewidth=0.2)
```
